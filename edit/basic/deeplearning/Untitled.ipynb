```python
plt.style.use('classic') # 스타일 정하기
x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x, np.sin(x),'--g') # -는 선, - 점선, -. 대쉬점선, : 은 동글점선, o는 그냥 점, g는 초록색,
# fig.savefig('my_figure.png') # 저장하기
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
plt.xlim(-1,11) # axis  설정하기
plt.ylim(-1.5,-1.5)
plt.axis([-1,11,-1.5,-1.5]) # 위와 동일하게 축설정
plt.axis('tight') # 타이트하게 그래프를 만듬 , eqaul을 쓸경우 가로세로 비율을 균등하게 설정 .
```

```python
plt.plot(x, np.sin(x))
plt.title("사인 함수") # 이름 붙이기
plt.xlabel("x") # 라벨 붙이기
plt.ylabel("sin(x)")
```

```python

plt.plot(x,np.sin(x), '-g', label='sin(x)')
plt.plot(x,np.cos(x), ':b', label='cos(x)')
plt.legend() # 범례를 생성
```

```python
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),marker,label="maker={0}".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0,1.8)
```

```python
rng =np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar() # 색상 척도 표시
```

```python
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x,y,yerr=dy,fmt='.k')
```
